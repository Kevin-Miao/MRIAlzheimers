# Dataset

### Variables and Context

One of the datasets is a cross-section (so horizontal cut of the brain) and the other data set (longitudinal) is a slice through the middle of the head.

### Cross-Sectional

```python
pd.read_csv('oasis_cross-sectional.csv').head(5)
```

**Variables**

*Demographics*
* **ID**: ID number (1-
* **M/F**: Gender (M/F)
* **Hand**: Handedness (L/R) (All people were right handed)
* **Age**
* **Education**: Years of Education
* **SES**: Social Economic Status by Hollingshead Index of Social Position (1: Highest Status, 5: Lowest Status)

*Clinical Data* (Actually has to do with Alzheimers)
* **MMSE:** Mini-Mental State Examination Score (0 = Worst, 30=Best)
Cutoffs for this test. 0-9: severe, 10-18: mildly impacted, 19-23: cognitive impairment, 23-30: healthy.
* **CDR:** Clinical Dementia Rating (Classified by a doctor) 
- 0 = Normal
- 0.5 = Very Mild Dementia
- 1 = Mild Dementia
- 2 = Moderate Dementia
- 3 = Severe Dementia

*MRI Data* (From imaging)
* **eTIv:** Estimated Intracranial Volume (in mm^3)
* **nWBV:** Normalized Whole-Brain Volume (as a percentage of all matter that is labeled as grey or white by software)
* **ASF:** Atlas Scaling Factor (Unitless) 
https://www.ncbi.nlm.nih.gov/pubmed/15488422
